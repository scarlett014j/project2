```python
import requests
import json
import pandas as pd
from config import api_key
```

```python
ord_url = f"https://www.thecocktaildb.com/api/json/v2/{api_key}/filter.php?c=Ordinary_Drink"
cocktail_url = f"https://www.thecocktaildb.com/api/json/v2/{api_key}/filter.php?c=Cocktail"
```

```python
response1 = requests.get(ord_url).json()
names = response1['drinks']
```

```python
ordinary = []
for name in names:
    ordinary.append(name['strDrink'])
    
ord_df = pd.DataFrame(ordinary)
ord_df.rename(columns = {0: "Name"}, inplace = True)
ord_df.head()
```

```python
ordinary2 = []
for name in names:
    ordinary2.append(name['idDrink'])
    
ord_df2 = pd.DataFrame(ordinary2)
ord_df2.rename(columns = {0: "ID"}, inplace = True)
ord_df2.head()
```

```python
ordinary3 = []
for name in names:
    ordinary3.append(name['strDrinkThumb'])
    
ord_df3 = pd.DataFrame(ordinary3)
ord_df3.rename(columns = {0: "Image URL"}, inplace = True)
ord_df3.head()
```

```python
ordinary_drinks = pd.concat([ord_df2, ord_df, ord_df3], axis =1)
ordinary_drinks
```

```python
response2 = requests.get(cocktail_url).json()
names2 = response2['drinks']
```

```python
cocktail = []
for name in names2:
    cocktail.append(name['strDrink'])
    
cocktail_df = pd.DataFrame(cocktail)
cocktail_df.rename(columns = {0: "Name"}, inplace = True)
cocktail_df.head()
```

```python
cocktail2 = []
for name in names2:
    cocktail2.append(name['idDrink'])
    
cocktail_df2 = pd.DataFrame(cocktail2)
cocktail_df2.rename(columns = {0: "ID"}, inplace = True)
cocktail_df2.head()
```

```python
cocktail3 = []
for name in names2:
    cocktail3.append(name['strDrinkThumb'])
    
cocktail_df3 = pd.DataFrame(cocktail3)
cocktail_df3.rename(columns = {0: "Image URL"}, inplace = True)
cocktail_df3.head()
```

```python
cocktail_drinks = pd.concat([cocktail_df2, cocktail_df, cocktail_df3], axis =1)
cocktail_drinks
```

```python
combined = [ordinary_drinks, cocktail_drinks]
combined_result = pd.concat(combined)
combined_result.set_index("ID")
```

```python
combined_result.to_csv("Drink Names and IDs.csv")
```

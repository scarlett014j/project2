```python
# Dependencies
import requests
import json
import pandas as pd

# Google developer API key
from config import gkey
```

```python
# geocoordinates
target_coordinates = "33.7490, -84.3880"
target_radius = 8000
target_type = "liquor_store"

# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "radius": target_radius,
    "type": target_type,
    "key": gkey
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
response = requests.get(base_url, params=params)
```

```python
print(response.url)
```

```python
# convert response to json
places_data = response.json()

# Print the json (pretty printed)
print(json.dumps(places_data, indent=4, sort_keys=True))
```

```python
print(places_data["results"][0]["geometry"]["location"])
print(places_data["results"][0]["name"])
```

```python
#print(places_data["results"][1])
```

```python
name = []
lat  = []
lng  = []
for i in places_data["results"]:
    name.append(i["name"])
    lat.append(i["geometry"]["location"]["lat"])
    lng.append(i["geometry"]["location"]["lng"])
```

```python
name
```

```python
locations_df = pd.DataFrame({"name":name, "lat":lat, "long":lng})
```
